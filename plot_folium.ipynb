```python
import requests
import folium
```

```python
data_path = "https://raw.githubusercontent.com/Gusb3ll/thailand-public-train-data/main/data/station/"
data_ext = ".json"
```

```python
arl = "arl"
bts = "bts"
bts_gold = "bts_gold"
mrt = "mrt"
mrt_blue = "mrt_blue"
mrt_purple = "mrt_purple"
```

```python
def showMap(line,location,zoom):
  data = requests.get(f"{data_path}{line}{data_ext}")
  map = folium.Map(location=location, zoom_start=zoom)
  for station in data.json():
    name_th = station['name']
    name_en = station['nameEng']
    marker = folium.Marker(
        location=[station['geoLat'],station['geoLng']],
        popup=f"{name_th} ({name_en})",
        tooltip=name_th)
    marker.add_to(map)
  return map
```

```python
showMap(arl,[13.736384, 100.636139],12)
```

```python
showMap(bts,[13.745409, 100.534108],11)
```

```python
showMap(bts_gold,[13.725378, 100.506535],15)
```

```python
showMap(mrt,[13.817077, 100.49366],12)
```

```python
showMap(mrt_blue,[13.766397, 100.505676],13)
```

```python
showMap(mrt_purple,[13.855414, 100.465164],13)
```
